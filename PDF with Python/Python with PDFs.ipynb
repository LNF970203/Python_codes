### PDF Rotation

```python
def PDFrotation(original_filename, new_filename, rotation):
    assert type(rotation) == int, "Rotation must be an integer"
    assert rotation >= 0 and rotation <= 360, "Rotation must be between 0 and 360"
    
    pdfFileobj = open(original_filename, 'rb')
    #creating the pdf reader
    pdfReader = PyPDF2.PdfFileReader(pdfFileobj)
    
    #creeating the pdf writer object
    pdfWriter = PyPDF2.PdfFileWriter()
    
    for page in range(pdfReader.numPages):
        
        pageobj = pdfReader.getPage(page)
        pageobj.rotateClockwise(rotation)
        
        #Adding the created page to the new pdf
        pdfWriter.addPage(pageobj)
        
        
    #new file object
    newfile = open(new_filename, 'wb')
    
    #writing rotated pages to new file
    pdfWriter.write(newfile)
    
    pdfFileobj.close()
    newfile.close()
```

```python
path1 = r"C:\Lachin\DataSets\PDF\case2.pdf"
path2 = r"C:\Lachin\DataSets\PDF\new_case3.pdf"
PDFrotation(path1,path2, rotation=180)
```
